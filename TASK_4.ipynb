```python
#load dataset
import pandas as pd
df=pd.read_csv("/content/traffic_accidents.csv")
df
```

```python
df.isnull().sum()
```

```python
#check null
df.replace(['UNKNOWN','UNAVAILABLE','OTHER','NOT APPLICABLE'], pd.NA, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Analyze Time of day patterns
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
sns.countplot(x='crash_hour',data=df)
plt.title("Crash Time of Day")
plt.xlabel("Crash Hour")
plt.ylabel("Count")
plt.show()
```

```python
#Analyze weather pattern
plt.figure(figsize=(15,10))
sns.countplot(x='weather_condition',data=df)
plt.title("Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#road surface condition
plt.figure(figsize=(15,10))
sns.countplot(x='roadway_surface_cond',data=df)
plt.title("Road Surface Condition")
plt.xlabel("Road Surface Condition")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#Analyze Lighting condition
plt.figure(figsize=(15,10))
sns.countplot(x='lighting_condition',data=df)
plt.title("Lighting Condition")
plt.xlabel("Lighting Condition")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#analyze road defective
plt.figure(figsize=(5,10))
sns.countplot(x='road_defect',data=df)
plt.title("Road Defect")
plt.xlabel("Road Defect")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python

#Analyze crash type and total injury
plt.figure(figsize=(6,5))
sns.countplot(x='crash_type',data=df)
plt.title("Crash Type")
plt.xlabel("Crash Type")
plt.ylabel("Count")
plt.show()
```

```python
#data frame of accident count by traffic control devices
traffic_control_device_counts = df['traffic_control_device'].value_counts()
print(traffic_control_device_counts)

#analyze traffic control device
plt.figure(figsize=(15,10))
sns.countplot(x='traffic_control_device',data=df)
plt.title("Traffic Control Device")
plt.xlabel("Traffic Control Device")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#analyze primary contributory causes under certain condions
weather_cause = df.groupby(['weather_condition', 'prim_contributory_cause']).size().unstack(fill_value=0)
lighting_cause = df.groupby(['lighting_condition', 'prim_contributory_cause']).size().unstack(fill_value=0)
road_cause = df.groupby(['roadway_surface_cond', 'prim_contributory_cause']).size().unstack(fill_value=0)

#print top 10 primary contributory cause
print("Top 10 Primary Contributory Causes:")
print(df['prim_contributory_cause'].value_counts().head(10))
```

```python
#analyze first_crash_type
plt.figure(figsize=(15,10))
sns.countplot(x='first_crash_type',data=df)
plt.title("First Crash Type")
plt.xlabel("First Crash Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#Analyze accident patterns by time and day

accident_counts_by_day = df['crash_day_of_week'].value_counts()
print("Accident Counts by Day of the Week:")
display(accident_counts_by_day)

injury_cols = ['injuries_total', 'injuries_fatal', 'injuries_incapacitating', 'injuries_non_incapacitating', 'injuries_reported_not_evident', 'injuries_no_indication']
severity_by_day = df.groupby('crash_day_of_week')[injury_cols].mean()
print("\nMean Injury Counts by Day of the Week:")
display(severity_by_day)

plt.figure(figsize=(10, 6))
accident_counts_by_day.sort_index().plot(kind='bar')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')
plt.title('Accident Counts by Day of Week')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
